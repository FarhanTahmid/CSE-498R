```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import easyocr
```

```python
def plotImage(image,title,cmap='gray'):
    plt.imshow(image,cmap)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
image=cv2.imread('Example Images/test.PNG')
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plotImage(gray_image,title="Grayscale Image")
```

```python
scaled_image = cv2.resize(gray_image, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
```

```python
final_image = cv2.fastNlMeansDenoising(scaled_image, None, 30, 7, 21)
```

```python
plotImage(final_image,title="Final Image")
```

```python
reader=easyocr.Reader(['en','bn'])
```

```python
result_from_text = reader.readtext(final_image)
```

```python
for (bbox, text, prob) in result_from_text:
    print(f'Text: {text}, Probability: {prob}')
```
